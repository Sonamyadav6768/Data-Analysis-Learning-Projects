```python
import seaborn as sns
import pandas as pd
import numpy as mp
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins['species'].value_counts()
```

```python
penguins['island'].value_counts()
```

```python
sns.scatterplot(data = penguins , x = "flipper_length_mm" , y = "body_mass_g", hue ="sex")
```

```python
sns.set_style("ticks")
sns.set_context("talk")

sns.scatterplot(data = penguins , x = "flipper_length_mm" , y = "body_mass_g", hue ="sex", palette="dark")
```

```python
sns.scatterplot(data=penguins, x ="species", y="body_mass_g", hue="island", style="sex")

```

```python
sns.stripplot(data=penguins, x ="species", y="body_mass_g", hue="island", dodge =True , jitter = True)
```

```python
sns.swarmplot(data=penguins, x ="species", y="body_mass_g", hue="island")
```

```python
sns.histplot(data=penguins, x="body_mass_g", hue="sex")
```

```python
sns.regplot(data=penguins,  x="body_mass_g", y="flipper_length_mm")
```

```python
sns.lineplot(data=penguins,  x="body_mass_g", y="flipper_length_mm", hue="island", style="sex")
```

```python
sns.jointplot(data=penguins,  x="body_mass_g", y="flipper_length_mm")
```

```python
sns.jointplot(data=penguins,  x="body_mass_g", y="flipper_length_mm", hue="sex")
```

```python
sns.jointplot(data=penguins,  x="body_mass_g", y="flipper_length_mm", hue="sex", kind="kde")
```

```python
sns.barplot(data=penguins, x = "species", y="body_mass_g" , hue="sex")
```

```python
sns.countplot(data=penguins , x="species")
```

```python
sns.boxplot(data=penguins, x = "species", y = "body_mass_g", hue="island")
```

```python
sns.violinplot(data=penguins, x = "species", y = "body_mass_g", hue="island", split=True)
```

```python
sns.violinplot(data=penguins, x = "species", y = "body_mass_g")
sns.swarmplot(data=penguins, x = "species", y = "body_mass_g")
```

```python
sns.kdeplot(data=penguins, x="body_mass_g",hue="species", fill=True)
```

```python
columns= ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
sns.heatmap(data=penguins[columns].corr())
```

```python
sns.heatmap(data=penguins[columns].corr(), annot=True)
```

```python
sns.rugplot(data=penguins, x = "species", y = "body_mass_g", height=0.3)
```

```python
sns.pairplot(data=penguins, hue="sex", diag_kind="kde")
```

```python
graph = sns.PairGrid(data=penguins, hue="sex", palette="Set2")
graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)
graph.map_offdiag(sns.barplot)
```
