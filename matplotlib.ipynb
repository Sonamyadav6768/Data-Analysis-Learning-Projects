```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data= {
    "salary" : [2000 , 3000 , 40000 , 49999 , 5000 , 22024, 42053],
    "dept" :  ['hr', 'it', 'finacne', 'it', 'hr', 'it', 'hr']
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.info()
```

```python
count = df["dept"].value_counts()
count
```

```python
plt.pie(count, labels=count.index , autopct= "% 1.2f", explode=[0 , 0.1 , 0.2])
plt.axis("equal")
plt.show()
```

```python
plt.bar(count, count.index)
plt.show()
```

```python
df["age"] = [22,54,67,3,86,25,8]  # same length as df
```

```python
plt.bar(df["age"], df["salary"], color = "red")
plt.grid()
plt.show()
```

```python
hr_sal = df[df["dept"]=="hr"]["salary"]
hr_sal
```

```python
it_sal = df[df["dept"]=='it']["salary"]
it_sal
```

```python
finance_sal = df[df["dept"]=="finacne"]["salary"]
finance_sal
```

```python
plt.boxplot([hr_sal, it_sal, finance_sal], labels=["hr", "IT", "fINANCE"])
plt.show()
```

```python
salary_sum = df.groupby("dept")["salary"].sum()
salary_sum
```

```python
plt.pie(salary_sum, labels=salary_sum.index , autopct = "%1.2f")
plt.show()
```

```python
hr_mean = sum(hr_sal)/len(hr_sal)
it_mean = sum(it_sal)/len(it_sal)
finance_mean = sum(finance_sal)/len(finance_sal)
```

```python
plt.bar(["hr", "it", "finance"],[hr_mean, it_mean, finance_mean])
plt.show()
```
