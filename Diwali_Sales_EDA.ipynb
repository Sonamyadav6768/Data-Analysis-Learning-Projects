```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/Diwali Sales Data.csv", encoding='unicode_escape')
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["Status" , "unnamed1"], axis=1, inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df["Amount"]= df["Amount"].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status': 'Shadi'})
```

```python
df.describe()
```

```python
df[["User_ID", "Age"]].describe()
```

```python
ax= sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
sns.barplot(x='Age Group', y="Amount", data=sales_age)
```

```python
ax = sns.countplot(x="Age Group", data = df , hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
state_data = df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by="Orders", ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = state_data, x="State" , y="Orders")
```

```python
state_data = df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by="Amount", ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = state_data, x="State" , y="Amount")
```

```python
ax= sns.countplot(data=df, x= "Marital_Status")
for bars in ax.containers:
    ax.bar_label(bars)

```

```python
state_data = df.groupby(['Marital_Status', "Gender"], as_index= False)['Amount'].sum().sort_values(by="Amount", ascending=False)
sns.set(rc={'figure.figsize':(4,5)})
sns.barplot(data = state_data, x="Marital_Status" , y="Amount", hue="Gender")
```

```python
ax= sns.countplot(data=df, x= "Occupation")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
state_data = df.groupby(['Occupation'], as_index= False)['Amount'].sum().sort_values(by="Amount", ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = state_data, x="Occupation" , y="Amount")
```
